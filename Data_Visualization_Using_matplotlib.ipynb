```python
#subplot

import numpy as np
a = np.array([13,33,553,53,63,23])
b = np.array([2,222,32,22,52,62])
c = np.array([7,1,27,37,477,57])
d = np.array([3,45,555,65,75,8])
```

```python
plt.plot(a,b)
```

```python
plt.subplot(4,3,1)
plt.plot(a,c, color="red")

plt.subplot(4,3,2)
plt.plot(a,d, color="red")

plt.subplot(4,3,3)
plt.plot(a,b, color="red")

plt.subplot(4,3,4)
plt.plot(b,c, color="blue")

plt.subplot(4,3,5)
plt.plot(b,d, color="blue")

plt.subplot(4,3,6)
plt.plot(b,a, color="blue")

plt.subplot(4,3,7)
plt.plot(c,a, color="green")

plt.subplot(4,3,8)
plt.plot(c,b, color="green")

plt.subplot(4,3,9)
plt.plot(c,d, color="green")

plt.subplot(4,3,10)
plt.plot(d,a, color="black")

plt.subplot(4,3,11)
plt.plot(d,b, color="black")

plt.subplot(4,3,12)
plt.plot(d,c, color="black")
```
